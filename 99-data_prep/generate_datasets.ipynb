You can download the dataset from here: https://www.kaggle.com/datasets/asaniczka/top-spotify-songs-in-73-countries-daily-updated/data

```python
import pandas as pd
from datetime import datetime
```

```python
data = pd.read_csv('universal_top_spotify_songs.csv')
data
```

```python
data_song_properties = data.drop(["daily_rank", "daily_movement", "weekly_movement", "country", "album_name", "album_release_date"], axis=1)
data_song_properties["snapshot_date"] = data_song_properties["snapshot_date"].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
data_song_properties = data_song_properties.sort_values(by="snapshot_date", ascending=True)
data_song_properties
```

```python
data_song_properties_latest = data_song_properties.drop_duplicates(subset=['spotify_id'], keep='first')
data_song_properties_latest
```

```python
data_song_details = data.drop(["popularity", "is_explicit", "duration_ms", "danceability", "energy", "key", "loudness", "mode", "speechiness", "acousticness", "instrumentalness", "liveness", "valence", "tempo", "time_signature"], axis=1)
data_song_details["snapshot_date"] = data_song_details["snapshot_date"].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
data_song_details
```

```python
data_song_properties.to_parquet('song_properties.parquet', index=True)
data_song_details.to_parquet('song_rankings.parquet', index=True)
```
